---
## 3. Filtering

### Observations
df_normal represents the dataset excluding Feed Change Events and Catalyst Replacements. These are planned downtime activities that affect process conditions. Removing them gives a more representative view of normal steady-state operation.
df_events represent all the records that do have an event. For later use in determining the impact of the events on the process conditions.

```python
df_normal = df_fcc[
    (df_fcc['Feed_Change_Event'] == 0)
    &
    (df_fcc['Catalyst_Replacement'] == 0)
]
```

```python
df_events = df_fcc[
    (df_fcc['Feed_Change_Event'] == 1)
    |
    (df_fcc['Catalyst_Replacement'] == 1)
]
```

---
## 4. Product Yield Comparison

### Observations
Comparison between the normal and events datasets shows little to no difference in Product Yield. This is likely caused by the synthetic nature of the dataset, which does not realistically simulate the impact of events on yield.

```python
df_prod_yield_normal = df_normal['Product_Yield'].agg(['mean', 'median', 'std'])
df_prod_yield_events = df_events['Product_Yield'].agg(['mean', 'median', 'std'])
pd.concat([df_prod_yield_normal, df_prod_yield_events], axis=1, keys=['PY_Normal', 'PY_Events'])
```

---
## 4. Sorting

## Observations
Sorting on max Product_Yield and Conversion_Rate shows a hard cap on 60 and 90. This is because of the nature of the synthetic dataset.

```python
df_fcc['Timestamp'] = pd.to_datetime(df_fcc['Timestamp'])
```

```python
df_normal_min_yield = df_normal.sort_values(by=['Product_Yield']).head(n=10)
df_normal_min_yield.sort_values(by=['Timestamp'])
```

```python
df_normal_max_yield = df_normal.sort_values(by=['Product_Yield'], ascending=False).head(n=100)
df_normal_max_yield.sort_values(by=['Timestamp'])
```

---
## 5. Ranking

## Observations
Ranking the Product Yield shows the limitations of the synthetically generated dataset. The max en min values of the Product_Yield are a hard limit, and thus occur many times. It does not show specific conditions that lead to optimum or minimum process conditions.

```python
df_normal['Yield_Rank'] = df_normal['Product_Yield'].rank(ascending=False)
df_normal.sort_values(by='Yield_Rank')
```

```python
df_normal['Yield_Rank'] = df_normal['Product_Yield'].rank(ascending=True, method='min')
df_normal.sort_values(by='Yield_Rank')
```
