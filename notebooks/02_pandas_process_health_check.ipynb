## 2. Descriptive statistics

### Observations
Product Yield and Conversion Rate show a left-skewed distribution 
(mean < median), suggesting occasional low-yield events such as 
process shutdowns or feed changes pulling the mean down.

```python
df_fcc.select_dtypes('float64').describe()
```

### Observations
Feed changes and catalyst replacements occur frequently — 9,979 and 4,358 events respectively over one year. Reducing these events could improve overall process efficiency and stabilise Product Yield and Conversion Rate.
