## 0. Imports

```python
import numpy as np

# Display settings
np.set_printoptions(precision=2, suppress=True)
```

---
## 1. Data Loading & Structure

Load the dataset as a NumPy array and inspect its basic structure.  
As a process operator: *how many tags, how many measurements?*

```python
data = np.genfromtxt("../data/fcc_dataset_large.csv", delimiter=",",
                     skip_header=1,
                     usecols=range(1, 27))
with open("../data/fcc_dataset_large.csv", "r") as f:
    data_headers = f.readline().split(sep=",")
data_headers = data_headers[1:-1]
```

**Observations:**  
*fcc_dataset_large.csv is a 2d array. Size is 105120 rows by 26 columns.
Because first and last columns are other datatypes than float, they are excluded.
Headers are separate in data_headers.*

---
## 2. Descriptive Statistics

Calculate mean, median and standard deviation per sensor (column).  
Key question: *where does mean ≠ median, and what might that mean operationally?*

```python
data_means = np.mean(data, axis=0)
data_medians = np.median(data, axis=0)
data_sds = np.std(data, axis=0)
for data_header, data_mean, data_median, data_sd in zip(data_headers,
                                  data_means,
                                  data_medians,
                                  data_sds):
    print(f"{data_header}:\n"
          f"\tmean :{data_mean:.2f}\n"
          f"\tmedian: {data_median:.2f}\n"
          f"\tstandard deviation: {data_sd:.2f}\n")
```

**Observations:**  
*Product_Yield and Conversion_Rate have mean < median. Probably caused by downtime or 
 process disturbances pulling the mean downward.*

---
## 3. Outlier Detection

Use percentiles and IQR to identify which sensors have the most outliers.  
Operational context: *outliers here are not data errors — they may be real process events.*

```python
data_q1s = np.percentile(data, 25, 0)
data_q3s = np.percentile(data, 75, 0)
iqrs = data_q3s - data_q1s
lower_limits = data_q1s - 1.5 * iqrs
upper_limits = data_q3s + 1.5 * iqrs
low_outliers = (data < lower_limits)
upper_outliers = (data > upper_limits)

col_width = 30

print(f"{'Sensor':<{col_width}} {'Mean':>10} {'Median':>10} {'Std':>10} {'Low':>8}"
      f"{'High':>8}")
print("-" * 82)

for header, mean, median, std, low, high in zip(data_headers, data_means,
                                               data_medians, data_sds,
                                               low_outliers.sum(axis=0),
                                               upper_outliers.sum(axis=0)):
    print(f"{header:<{col_width}} {mean:>10.2f} {median:10.2f} {std:>10.2f} {low:>8}"
          f"{high:>8}")

```

**Observations:**  
*Because of the nature of the dataset (synthetically generated). There are no real outliers
in the data. Except for Reward_Score (possibly Reinforcement learning signal),
Feed_Change_Event and Catalyst_Replacement (binary events).*
