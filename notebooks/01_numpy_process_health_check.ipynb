## 1. Data Loading & Structure

Load the dataset as a NumPy array and inspect its basic structure.  
As a process operator: *how many tags, how many measurements?*

```python
data = np.genfromtxt("../data/fcc_dataset_large.csv", delimiter=",",
                     skip_header=1,
                     usecols=range(1, 27))
with open("../data/fcc_dataset_large.csv", "r") as f:
    data_headers = f.readline().split(sep=",")
data_headers = data_headers[1:-1]
```

**Observations:**  
*fcc_dataset_large.csv is a 2d array. Size is 105120 rows by 26 columns.
Because first and last columns are other datatypes than float, they are excluded.
Headers are separate in data_headers.*

---
## 2. Descriptive Statistics

Calculate mean, median and standard deviation per sensor (column).  
Key question: *where does mean ≠ median, and what might that mean operationally?*

```python
data_means = np.mean(data, axis=0)
data_medians = np.median(data, axis=0)
data_sds = np.std(data, axis=0)
for data_header, data_mean, data_median, data_sd in zip(data_headers,
                                  data_means,
                                  data_medians,
                                  data_sds):
    print(f"{data_header}:\n"
          f"\tmean :{data_mean:.2f}\n"
          f"\tmedian: {data_median:.2f}\n"
          f"\tstandard deviation: {data_sd:.2f}\n")
```

**Observations:**  
*Product_Yield and Conversion_Rate have mean < median. Probably caused by downtime or 
 process disturbances pulling the mean downward.*
